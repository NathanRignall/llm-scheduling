```python
# Import required dependencies
import matplotlib.pyplot as plt
import numpy as np

# Configure matplotlib
plt.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.Set1(np.linspace(0, 1, 10)))
```

```python
# plot dumps on histogram
plt.figure(figsize=(10, 3))
plt.plot([0, 1], [0, 1], label="Linear")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("../data/figures/sample.pdf", bbox_inches="tight", format="pdf")
```
